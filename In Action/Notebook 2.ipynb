# Data Cleaning

```
df_training = train_complete.copy()
```

```
df_training['bad_sleep'] = np.where(df_training['sleep_quality'] == 'poor', 1, 0)
df_training = df_training.drop(columns=['exam_difficulty','gender','sleep_quality'])
df_training = pd.get_dummies(df_training, columns=['course','study_method','facility_rating'],dtype=int)
df_training['exam_score'] =df_training.pop('exam_score')
```

```
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Ensure X and y are properly defined based on the main dataset
# Assuming df_dropped_id contains all features and 'exam_score' is the target
X = df_dropped_id.drop('exam_score', axis=1)
y = df_dropped_id['exam_score']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

XGB = XGBRegressor()
XGB.fit(X_train, y_train)
y_pred = XGB.predict(X_test)

# Calculate and print regression metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
reg_model1 = RandomForestRegressor(n_estimators = 10, random_state = 0)
reg_model2 = XGBRegressor(random_state=0)
reg_model4 = MLPRegressor(random_state=0, max_iter=2000, tol=0.1)
reg_model5 = GradientBoostingRegressor(random_state=0)
reg_model6 = LinearRegression()
reg_model8 = Ridge(alpha=1.0)
reg_model9 = linear_model.Lasso(alpha=0.1)
reg_model7 = VotingRegressor(estimators=[('gb', reg_model5), ('rge', reg_model8)]) # ('rf',reg_model1),('xgb', reg_model2),('mp',reg_model3),('mlp',reg_model4),
```

```python
model_list = [reg_model1, reg_model2, reg_model4, reg_model5,reg_model6,
              reg_model7, reg_model8, reg_model9]
```

```python
X = df_model.drop('SalePrice', axis=1)
y = df_model['SalePrice']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
def train_regression_models(model_list, X_train, X_test, y_train, y_test):
    for model in model_list:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        mae = mean_absolute_error(y_test, y_pred)
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
        cleaned_model_name = re.sub(r'\([^)]*\)', '', str(model))
        print(cleaned_model_name)
        print(f"Mean Absolute Error: {mae}")
        print(f"Mean Squared Error: {mse}")
        print(f"R-squared: {r2}")
        print(" ")
        joblib.dump(model, f'{cleaned_model_name}.pkl')
```

```python
train_regression_models(model_list, X_train, X_test, y_train, y_test)
```
